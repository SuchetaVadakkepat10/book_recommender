```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
axis= plt.axes()
sns.heatmap(books.isna().transpose(), cbar = False, ax=axis)
plt.xlabel("Columns")
plt.ylabel("Missing values")
plt.show()
```

```python
import numpy as np
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
# books["age"] = 2025 - books["publication_year"]

books.columns
```

```python
columns_of_interest = ["num_pages", "published_year", "average_rating", "missing_description"]

correlation_matrix = books[columns_of_interest].corr(method='spearman')
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Selected Features")
plt.show()
```

```python
# Lesser Correlation between num_pages and average_rating
books_no_missing = books[ ~books["missing_description"].isna() &
      ~books["num_pages"].isna()&
      ~books["published_year"].isna() &
    ~books["average_rating"].isna()]
books_no_missing 
```

```python
books_no_missing["words_in_description"] = books_no_missing["description"].str.split().str.len()
books_no_missing
```

```python
books_no_missing.loc[books_no_missing["words_in_description"].between(25,34), "description"] 
```

```python
book_missing_25 = books_no_missing[books_no_missing["words_in_description"] >= 25]
```

```python
book_missing_25["title_and_subtitle"] = np.where(book_missing_25["subtitle"].isna(),
                                                  book_missing_25["title"], 
                                                  book_missing_25[["title", "subtitle"]].astype(str).agg(': '.join, axis=1))

book_missing_25
```

```python
book_missing_25.columns
```

```python
book_missing_25["tagged_description"] = book_missing_25[["isbn13", "description"]].astype(str).agg(' '.join, axis=1)
```

```python
book_missing_25
```

```python
{
    book_missing_25.drop(columns=["subtitle","words_in_description", "missing_description"], axis=1).to_csv("books_cleaned.csv", index=False)

}
```
